```python
import os,re,csv
def showInfo(folderPath):
  dataDict = dict()
  sentence,wordCount = 0,0
  for i in os.listdir(folderPath):
    f = open(folderPath+"/"+i,"r")
    for i in f.readlines()[2:]:
      sentence += len(re.findall('৷|।|\?|!',i))
      for j in re.split('।|,|\s+|-|\?|\(|–|“|’|‘|;|!|\)|\'|\"|৷|`|”|\n',i):
        j = re.sub('\.*',"",j)
        if len(j) > 1:
          wordCount += 1
          if j in dataDict.keys():
            dataDict[j] += 1
          else: dataDict[j] = 1
    f.close()
   

  i = 1
  for element in dataDict:
    print(f"{i}. {element}   -   {dataDict[element]}")
    i += 1

  print(f"\n\n\t\t{re.split('/',folderPath)[-1]}")
  print(f"Word = {wordCount}")
  print(f"Sentence = {sentence}\n")



```

```python
showInfo("/content/Stories Script")
```
